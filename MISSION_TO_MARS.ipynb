## SETTING

```python
#import Beautifulsoup
from bs4 import BeautifulSoup as bs
from splinter import Browser
```

```python
#set the executable path and initialize the chrome browser
executable_path = {'executable_path': './chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

## Visit the NASA web sites

```python
#visit  the mars nasa new site
url='https://mars.nasa.gov/news/'
browser.visit(url)
```

```python
html = browser.html
news_soup = bs(html, 'html.parser')

#print(news_soup)

slide_element = news_soup.select_one('ul.item_list li.slide')
#print(slide_element)
```

```python
slide_element.find('div', class_="content_title")
```

```python
#use the parent element to find the first a tag and save it as news-title
news_title=slide_element.find('div',class_="content_title").get_text()
news_title
```

```python
news_paragraph=slide_element.find('div',class_="article_teaser_body").get_text()
news_paragraph
```

## JPL SPACE IMAGES FEATURED IMAGE

```python
#visit url
url='https://www.jpl.nasa.gov/spaceimages/?search-&category-Mars'
browser.visit(url)
```

```python
#asking splinter to go on the site hit a button with class name full_image
full_image_button=browser.find_by_id('full_image')
full_image_button.click()
```

```python
#Find the more info button and click that
browser.is_element_present_by_text('more info',wait_time=1)
more_info_element=browser.find_link_by_partial_text('more info')
more_info_element.click()
```

```python
#parse the results html with soup
html=browser.html
image_soup=bs(html, 'html.parser')
```

```python
img_url=image_soup.select_one('figure.lede a img').get('src')
img_url
```

```python
#use the base url to create an absolute url
img_url=f'https://www.jpl.nasa.gov{img_url}'
img_url
```

## MARS WHEATHER

```python
executable_path = {'executable_path': './chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url='https://twitter.com/marswxreport?lang=en'
browser.visit(url)
```

```python
html=browser.html
weather_soup=bs(html, 'html.parser')
```

```python
#fisrt find a tweet with the data-name 'Mars weather'
mars_weather_tweet=weather_soup.find('div',
                                    attrs={
                                        'class':"tweet",
                                        'data-name':'Mars Weather'
                                        })
```

```python
mars_weather=mars_weather_tweet.find('p','tweet-text').get_text()
mars_weather
```

```python
executable_path = {'executable_path': './chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url='https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k=target&v1=Mars'
browser.visit(url)
```

```python
hemisphere_image_urls=[]

#first get a list of all the hemispheres

links=browser.find_by_css('a.product-item h3')
print(links)
for item in range(len(links)):
    hemisphere={}

    #we have to find the element on each loop to avoid a stale element exception
    browser.find_by_css('a.product-item h3')[item].click()

    #next we find the sample image anchor tag and extract the href
    sample_element=browser.find_link_by_text('Sample').first
    hemisphere['img_url']=sample_element['href']

    #get hemisphere title
    hemisphere['title']=browser.find_by_css('h2.title').text

    #Append hemisphere object to list
    hemisphere_image_urls.append(hemisphere)

    #finally we navigate 
    browser.back()
```

```python
hemisphere_image_urls
```
